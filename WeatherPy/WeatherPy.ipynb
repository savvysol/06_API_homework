# Observable Trends

    By selecting 600 cities randomly and gathering current weather data on each one, we can make some general assumptions about over all weather patterns.

## Temperature

    Our data clearly reveals that the closer you are to the equator (0 degrees latitdue) the higher the temperature.  The average temperature was 66.01 degrees.

## Humidity

    Humidity did not appear to be connected to latitude.  It was interesting to note that more places have a higher humidity rating (average of 71.51%) than lower.

## Cloudiness

    Cloudiness was interesting in that many cities appeared to "group" up at certain cloudiness ratings.  Specifically, 100%, 76%, 40% and 0%.  

## Wind Speed

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
from time import gmtime, strftime
from pprint import pprint
import gmaps

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []
countries = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=2500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    country = citipy.nearest_city(lat_lng[0], lat_lng[1]).country_code
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        countries.append(country)

# Print the city count to confirm sufficient count.
rand_count = len(lats)
city_pair = len(cities)

print(f"We tried {rand_count} pairs of coordinates and found {city_pair} cities that match.")
```

```python
# Put the Cities in a DataFrame
raw_cities = pd.DataFrame({'City':cities,
                        'Country':countries})
raw_cities.head()
```

```python
base_url = 'http://api.openweathermap.org/data/2.5/weather?'
json_url = f"{base_url}appid={api_key}&q={raw_cities['City'][48]},{raw_cities['Country'][48]}"

response = requests.get(json_url).json()
# pprint(response)
```

```python
name = response['name']
ccode = response['sys']['country']
city_country = f"{name}, {ccode}"
humidity = response['main']['humidity']
temp = response['main']['temp']
clouds = response['clouds']['all']
wind = response['wind']
lat = response['coord']['lat']
long = response['coord']['lon']

print(f"""
      The Name is: {city_country} @ {lat}, {long}
      Humidity: {humidity}
      Temperature: {temp}
      Clouds: {clouds}
      Wind: {wind}
      """)
```

```python
Cities = raw_cities.dropna()
Cities.head()
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# initialize count variables
success = 0
whoops = 0

# record the time of the pull
today = strftime("%a, %b %d, %Y", gmtime())

# set the loop to call in my cities
for index, row in Cities.iterrows():
    
    # Build the URL
    base_url = 'http://api.openweathermap.org/data/2.5/weather?'
    
    city_country = f"{row['City']},{row['Country']}"
    
    json_url = f"{base_url}appid={api_key}&q={city_country}"
    
    # Make the call
    response = requests.get(json_url).json()
      
       
    # populate the data frame
    try:
        Cities.loc[index,'Name'] = city_country
        Cities.loc[index,'Lat'] = response['coord']['lat']
        Cities.loc[index,'Long'] = response['coord']['lon']
        Cities.loc[index,'Lat, Long'] = f"{row['Lat']},{row['Long']}"
        Cities.loc[index,'Temp'] = response['main']['temp']
        Cities.loc[index,'Max Temp'] = response['main']['temp_max']
        Cities.loc[index,'Humidity'] = response['main']['humidity']
        Cities.loc[index,'Clouds'] = response['clouds']['all']
        Cities.loc[index,'Wind Speed'] = response['wind']['speed']
        Cities.loc[index,'Wind Dir'] = response['wind']['deg']
        success += 1
        print(f'  {index+1}/{city_pair} | Found:  {success} | Success Rate: {round(success/(success+whoops)*100,1)}% | Complete: {round((index+1)/city_pair*100,1)}% | {city_country} ')
        if success == 600:
            break
    except Exception as e: 
        whoops += 1
        print(f"* {index+1}/{city_pair} | Failed: {whoops} | Failure Rate: {round(whoops/(index+1)*100,1)}% | Complete: {round((index+1)/city_pair*100,1)}% | {city_country} ")

        
```

```python
# Print a summary Success Statement
print(f"""
------------------------------------------------------
On {today} We looked at {city_pair} cities assembled by CitiPy
and ran those through the OpenWeather API.

{success} of the {city_pair} cities were found.
{whoops} gave an error and will be dropped.

Our success rate was {round(success/(success+whoops)*100,3)}%
we were successful with {success} of them.
-----------------------------------------------------
""")


    
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
clean_cities = Cities.sort_values('Lat').dropna()

# Convert Temp K to F
clean_cities['Temp F'] = round((clean_cities['Temp']-273.15) * 9/5 + 32,2)

print(len(clean_cities))
clean_cities.head()


```

```python
# Set the variables
lats = clean_cities['Lat']
temp = clean_cities['Temp F']
humid = clean_cities['Humidity']
clouds = clean_cities['Clouds']
wind = clean_cities['Wind Speed']

# Some observable stats
mean_temp = round(sum(temp)/len(temp),2)
mean_humidity = round(sum(humid)/len(humid),2)
mean_clouds = round(sum(clouds)/len(clouds),2)
mean_wind = round(sum(wind)/len(wind),2)

print(mean_temp)
print(mean_humidity)
print(mean_clouds)
print(mean_wind)
```
